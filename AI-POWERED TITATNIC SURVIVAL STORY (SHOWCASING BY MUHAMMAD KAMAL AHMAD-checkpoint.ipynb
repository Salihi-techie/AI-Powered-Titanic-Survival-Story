# My AI-Powered Titanic Survival Story

### My Dataq Journey: Learning to Load Data Seamlessly

When I started this project, I received the Titanic dataset in an Excel file, I faced some initial challenges trying to directly load it into my Jupyter Notebook. It was a bit frustrating! However, through this process, and with some help from AI tools like Google Gemini, I learned the importance and ease of converting such files to a CSV format. this simple step made data loading much smoother using Pandas and taught me a valuable lesson in data preparation - getting the data in the right format is key!

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os # Just in case for directory checks

sns.set_style("whitegrid")

# IMPORTANT: Make sure this matches the exact name you saved:
file_path_csv = 'titanic_final.csv' 

df = None 

try:
    print(f"Attempting to read '{file_path_csv}' as a raw CSV (single column initially)...")
    # Read the file line by line, without assuming a delimiter initially
    # Use 'header=None' to ensure no row is skipped assuming it's a header.
    df_raw = pd.read_csv(file_path_csv, header=None, encoding='utf-8')
    print("Raw CSV loaded. Now attempting to split the single column.")

    # If df_raw has only one column, we proceed with manual splitting
    if df_raw.shape[1] == 1:
        single_column_name = df_raw.columns[0]
        split_data = df_raw[single_column_name].str.split(',', expand=True)

        # The first row of `split_data` contains our actual column names
        column_names = split_data.iloc[0]

        # Assign these names as the new DataFrame's column headers
        df = split_data[1:] # All rows from the second row onwards (skipping header row)
        df.columns = column_names
        
        # Convert numeric columns from object to numeric types where possible
        for col in ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce') # 'coerce' turns unconvertible values into NaN
        
        print("Successfully split single column into multiple columns!")
    else:
        # This case would mean pd.read_csv somehow worked directly, which is unexpected but good.
        df = df_raw
        print("Dataset loaded successfully with multiple columns directly (unexpected but good!).")

except UnicodeDecodeError:
    print(f"UTF-8 failed for '{file_path_csv}' during raw read. Trying with latin1 encoding...")
    df_raw = pd.read_csv(file_path_csv, header=None, encoding='latin1')
    
    if df_raw.shape[1] == 1:
        single_column_name = df_raw.columns[0]
        split_data = df_raw[single_column_name].str.split(',', expand=True)
        column_names = split_data.iloc[0]
        df = split_data[1:]
        df.columns = column_names

        for col in ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')
        
        print("Successfully split single column with latin1 encoding.")
    else:
        df = df_raw
        print("Dataset loaded successfully with multiple columns directly using latin1.")

except FileNotFoundError:
    print(f"Error: The file '{file_path_csv}' was not found.")
    print(f"Please make sure you saved it as '{file_path_csv}' in the correct directory, or provide the full path.")
    df = None 
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None 

if df is None:
    print("\nFATAL ERROR: Dataset could not be loaded and processed into multiple columns. Please double-check your file or consider getting a fresh copy of the Titanic CSV from a reliable source (like Kaggle's Titanic competition dataset).")
    import sys
    sys.exit("Exiting due to critical data loading failure.")

# Display the first few rows of the DataFrame to confirm it loaded correctly
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Display basic information about the DataFrame
print("\nDataFrame Info:")
df.info()

# Check column names
print("\nDataFrame Columns:")
print(df.columns)
```

## Question 1: What is the survival rate of passengers?

```python
# --- Business Question 1: What is the survival rate of passengers? ---

# Calculate the survival rate
# Ensure 'survived' column is numeric, which we already did in the loading step
# but it's good practice to ensure before calculations.
df['survived'] = pd.to_numeric(df['survived'], errors='coerce')
survival_rate = df['survived'].mean() * 100
print(f"\n1. Overall Survival Rate: {survival_rate:.2f}%")

# Visualize the survival count
plt.figure(figsize=(7, 5))
sns.countplot(x='survived', data=df, palette='viridis')
plt.title('Passenger Survival Count (0 = Did Not Survive, 1 = Survived)')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualize the survival rate as a percentage with a bar plot
plt.figure(figsize=(7, 5))
non_survival_rate = (1 - survival_rate / 100) * 100
survival_data = pd.DataFrame({
    'Status': ['Did Not Survive', 'Survived'],
    'Percentage': [non_survival_rate, survival_rate]
})
sns.barplot(x='Status', y='Percentage', data=survival_data, palette='coolwarm')
plt.title('Overall Survival Rate Percentage')
plt.ylabel('Percentage of Passengers (%)')
plt.ylim(0, 100) # Ensure y-axis goes from 0 to 100
for index, row in survival_data.iterrows():
    plt.text(index, row['Percentage'] + 2, f'{row["Percentage"]:.2f}%', ha='center', va='bottom', fontsize=10)
plt.show()
```

## Question 2: What is the gender distribution of passengers?

```python
# --- Business Question 2: What is the gender distribution of passengers? ---

# Get the count of passengers by gender
gender_distribution = df['sex'].value_counts()
print("\n2. Gender Distribution of Passengers:")
print(gender_distribution)

# Calculate gender percentages
gender_percentage = df['sex'].value_counts(normalize=True) * 100
print("\nGender Distribution Percentage:")
print(gender_percentage.round(2))

# Visualize the gender distribution using a countplot
plt.figure(figsize=(7, 5))
sns.countplot(x='sex', data=df, palette='pastel')
plt.title('Distribution of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualize the gender distribution using a pie chart (good for proportions)
plt.figure(figsize=(8, 8))
plt.pie(gender_percentage, labels=gender_percentage.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Paired'))
plt.title('Percentage of Passengers by Gender')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Question 3: How does survival rate differ by class?

```python
# --- Business Question 3: How does survival rate differ by class? ---

# Calculate survival rate by passenger class
survival_by_class = df.groupby('class')['survived'].mean() * 100
print("\n3. Survival Rate by Passenger Class:")
print(survival_by_class.round(2))

# Visualize survival rate by class using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_by_class.index, y=survival_by_class.values, palette='coolwarm')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (%)')
plt.ylim(0, 100) # Ensure y-axis goes from 0 to 100
for index, value in enumerate(survival_by_class.values):
    plt.text(index, value + 2, f'{value:.2f}%', ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Another way to visualize using seaborn.catplot (facet grid for comparison)
# This shows counts for survived/not survived within each class
g = sns.catplot(x='class', hue='survived', data=df, kind='count', palette='muted', height=6, aspect=1.2)
g.set_axis_labels("Passenger Class", "Number of Passengers")
g.set_titles("Survival Count by Passenger Class")
g.set_xticklabels(["First", "Second", "Third"])
g._legend.set_title("Survived")
new_labels = ['No', 'Yes']
for t, l in zip(g._legend.texts, new_labels): t.set_text(l)
plt.suptitle('Survival Count by Passenger Class', y=1.02) # Adjust suptitle position
plt.show()
```

## Question 4: What is the distribution of passengers ages?

```python
# --- Business Question 4: What is the distribution of passengers ages? ---

# Check for missing values in 'age'
missing_age = df['age'].isnull().sum()
print(f"\n4. Number of missing values in 'age' column: {missing_age}")

# If there are missing values, you might consider imputing them (e.g., with the median)
# For this project, we'll proceed by dropping NaNs for visualization to keep it simple,
# but for robust analysis, imputation is often preferred.
# Let's fill missing ages with the median for a cleaner distribution plot
median_age = df['age'].median()
df['age_filled'] = df['age'].fillna(median_age)
print(f"Missing ages filled with median: {median_age:.1f}")

# Get basic statistics for age
print("\nAge Distribution Statistics:")
print(df['age_filled'].describe().round(2))

# Visualize the age distribution using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['age_filled'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualize the age distribution using a violin plot for a different perspective (shows density and quartiles)
# A box plot would also work, but violin plots give more detail on density
plt.figure(figsize=(8, 6))
sns.violinplot(y=df['age_filled'], color='lightgreen')
plt.title('Distribution of Passenger Ages (Violin Plot)')
plt.ylabel('Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Question 5: How many passengers embarked from each location?

```python
# --- Business Question 5: How many passengers embarked from each location? ---

# Check for missing values in 'embark_town'
missing_embark_town = df['embark_town'].isnull().sum()
print(f"\n5. Number of missing values in 'embark_town' column: {missing_embark_town}")

# Drop rows with missing 'embark_town' if there are only a few, as it's a categorical variable
if missing_embark_town > 0:
    df_cleaned_embark = df.dropna(subset=['embark_town']).copy()
    print(f"Dropped {missing_embark_town} rows with missing 'embark_town' for this analysis.")
else:
    df_cleaned_embark = df.copy() # No changes needed if no missing values

# Get the count of passengers by embarkation town
embark_counts = df_cleaned_embark['embark_town'].value_counts()
print("\nPassenger Count by Embarkation Town:")
print(embark_counts)

# Calculate embarkation percentages
embark_percentage = df_cleaned_embark['embark_town'].value_counts(normalize=True) * 100
print("\nEmbarkation Town Percentage:")
print(embark_percentage.round(2))

# Visualize the distribution using a countplot
plt.figure(figsize=(8, 6))
sns.countplot(x='embark_town', data=df_cleaned_embark, palette='viridis', order=embark_counts.index)
plt.title('Number of Passengers Embarked from Each Location')
plt.xlabel('Embarkation Town')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualize the distribution using a pie chart for percentages
plt.figure(figsize=(8, 8))
plt.pie(embark_percentage, labels=embark_percentage.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Percentage of Passengers by Embarkation Town')
plt.axis('equal')
plt.show()
```

## Question 6: Wat is the average fare paid per class?

```python
# --- Business Question 6: What is the average fare paid per class? ---

# Ensure 'fare' column is numeric
df['fare'] = pd.to_numeric(df['fare'], errors='coerce')

# Calculate the average fare paid per class
average_fare_by_class = df.groupby('class')['fare'].mean()
print("\n6. Average Fare Paid per Class:")
print(average_fare_by_class.round(2))

# Visualize the average fare per class using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=average_fare_by_class.index, y=average_fare_by_class.values, palette='viridis')
plt.title('Average Fare Paid per Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
# Adding value labels on top of bars
for index, value in enumerate(average_fare_by_class.values):
    plt.text(index, value + 2, f'${value:.2f}', ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# You could also use a box plot or violin plot to show the distribution of fares within each class
# (not just the average, but the spread)
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='fare', data=df, palette='magma')
plt.title('Fare Distribution by Passenger Class (Box Plot)')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.ylim(0, df['fare'].max() * 1.05) # Adjust y-limit to show outliers if any
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# --- Business Question 7: How does gender affect survival rates? ---

# Calculate survival rate by gender
survival_by_gender = df.groupby('sex')['survived'].mean() * 100
print("\n7. Survival Rate by Gender:")
print(survival_by_gender.round(2))

# Visualize survival rate by gender using a bar plot
plt.figure(figsize=(7, 5))
sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values, palette='coolwarm')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate (%)')
plt.ylim(0, 100) # Ensure y-axis goes from 0 to 100
for index, value in enumerate(survival_by_gender.values):
    plt.text(index, value + 2, f'{value:.2f}%', ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Another way to visualize using seaborn.catplot (shows counts for survived/not survived within each gender)
g = sns.catplot(x='sex', hue='survived', data=df, kind='count', palette='mako', height=6, aspect=1.2)
g.set_axis_labels("Gender", "Number of Passengers")
g.set_titles("Survival Count by Gender")
g._legend.set_title("Survived")
new_labels = ['No', 'Yes']
for t, l in zip(g._legend.texts, new_labels): t.set_text(l)
plt.suptitle('Survival Count by Gender', y=1.02)
plt.show()
```

## Question 8: What is the correlation between fare and survival?

```python
# --- Business Question 8: What is the correlation between fare and survival? ---

# Ensure 'fare' and 'survived' columns are numeric (already handled but good to reinforce)
df['fare'] = pd.to_numeric(df['fare'], errors='coerce')
df['survived'] = pd.to_numeric(df['survived'], errors='coerce')

# Calculate the average fare for survivors vs. non-survivors
average_fare_by_survival = df.groupby('survived')['fare'].mean()
print("\n8. Average Fare for Survivors vs. Non-Survivors:")
print(average_fare_by_survival.round(2))

# Visualize average fare by survival status
plt.figure(figsize=(7, 5))
sns.barplot(x=average_fare_by_survival.index, y=average_fare_by_survival.values, palette='viridis')
plt.title('Average Fare by Survival Status (0=No, 1=Yes)')
plt.xlabel('Survival Status')
plt.ylabel('Average Fare')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])
for index, value in enumerate(average_fare_by_survival.values):
    plt.text(index, value + 2, f'${value:.2f}', ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualize the distribution of fares for survivors and non-survivors using a violin plot
# This shows density and spread, not just the mean.
plt.figure(figsize=(8, 6))
sns.violinplot(x='survived', y='fare', data=df, palette='pastel')
plt.title('Fare Distribution by Survival Status (0=No, 1=Yes)')
plt.xlabel('Survival Status')
plt.ylabel('Fare')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])
plt.ylim(0, df['fare'].quantile(0.99) * 1.1) # Limit y-axis to focus on main distribution, hide extreme outliers
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# You could also calculate Pearson correlation coefficient for a numerical measure
correlation_fare_survival = df['fare'].corr(df['survived'])
print(f"\nPearson correlation between Fare and Survival: {correlation_fare_survival:.2f}")
```
